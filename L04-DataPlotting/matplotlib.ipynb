# Matplotlib
A library for plotting graphs

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([0,6,147,234])
ypoints = np.array([0,18,195,250])
```

```python
# print graph joining points
plt.plot(xpoints,ypoints, 'o')
plt.show()
```

```python
# print points
plt.plot(xpoints,ypoints, 'o')
plt.show()
```

```python
# print points as marker
plt.plot(xpoints,ypoints, marker = '*')
plt.show()
```

```python
# showing dotted lines
plt.plot(xpoints,lw="20",ms=30)
plt.show()
```

```python
# generate arrays for x and y
x1 = np.linspace(0,200,num = 50 )
y1 = np.random.randint(50,100, size=100).reshape((2, 50))

# set font properties for labels and title
font1 = { "family": "serif", "size": 21}
font2 = { "family": "monospace", "size": 18}

plt.xlabel("Unknown X", fontdict=font2)
plt.ylabel("Unknown Y", fontdict=font2)

plt.title("Main Title of Graph",fontdict=font1)

plt.plot(x1, y1[0])
plt.plot(x1, y1[1])

# set grid lines
plt.grid(axis='y', color="black", linestyle=":", linewidth=1.5)
plt.show()
```

```python
# display multiple plots 
x  = np.arange(100,step=5)
y1 =  np.random.randint(50,70,size=20)
y2 =  np.random.randint(20,30,size=20)

plt.subplot(2,1,1) # want 2rows and 1 column to stack on top of other
plt.plot(x,y1)

plt.subplot(2,1,2) # want 2rows and 1 column to stack on top of other
plt.plot(x,y2)

plt.show()
```

```python
# display multiple plots 
x  = np.arange(100,step=5)
y1 =  np.random.randint(50,70,size=20)
y2 =  np.random.randint(20,30,size=20)

plt.subplot(1,2,1) # want 1row and 2 column to place side by side
plt.title("First Graph")
plt.plot(x,y1)

plt.subplot(1,2,2)  # want 1row and 2 column to place side by side
plt.title("Second Graph")
plt.plot(x,y2)

# super title
plt.suptitle("My All Graphs")
plt.show()
```

```python
# create scatter plot

x  = np.linspace(0,100,num= 20)
y1 = np.random.randint(30,50, size= 20)
y2 = np.random.randint(30,50, size= 20)


plt.scatter(x,y1)
plt.scatter(x,y2)


plt.grid(axis= 'y')
plt.show()
```

```python
# scatter plot with each dot different color 

x  = np.linspace(0,100,num= 20)
y1 = np.random.randint(30,50, size= 20)
y2 = np.random.randint(30,50, size= 20)

colors = np.repeat(["red","yellow","green", "magenta", "orange"],4 )

plt.scatter(x,y1,c=colors)

plt.scatter(x,y2,c=np.flip(colors))


plt.grid(axis= 'y')
plt.show()
```

```python
# A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.
# define a colormap with cmap

x  = np.linspace(0,100,num= 20)
y1 = np.random.randint(30,50, size= 20)
y2 = np.random.randint(30,50, size= 20)

colors = np.random.randint(0,100,size=20) 

plt.scatter(x,y1,c=colors, cmap='viridis')

# show colorbar
plt.colorbar()

plt.scatter(x,y2,c=colors, cmap="Blues")

plt.colorbar()

plt.grid(axis= 'y')
plt.show()
```

```python
# creating dots with custom sizes
sizes = np.random.randint(20,1000,size=20)
colors = np.random.randint(0,100,size=20)

x  = np.linspace(0,100,num= 20)
y = np.random.randint(30,50,size= 20)

plt.scatter(x,y,s= sizes,c = colors , cmap= "nipy_spectral" , alpha=0.5)
plt.colorbar()
plt.show()
```

## Bar Graphs

```python
x = np.array(["2018","2019","2020","2021"]) 
y = np.array([10,30,25,45])
plt.bar(x,y, width=0.3)
plt.show()
```

```python
# Horizontal bars 

x = np.array(["2018","2019","2020","2021"]) 
y = np.array([10,30,25,45])
plt.barh(x,y, color="#4CAF50",  height=0.6)
plt.show()
```

## Histogram 
Histogram are graphs showing frequencies distributions. 

```python
x = np.random.normal(170,10,250) # normalized data
plt.hist(x)
plt.show()
```

## PIE Charts

```python
x = np.array([3,3,6])
mycolors = np.array(["red","purple","orange"])
mylabels = ["First","Second","Third"]
                  
plt.pie(x, colors= mycolors ,labels= mylabels , explode = [0,0.2,0.1], shadow = True )
plt.legend(title="Pie chart fields")
plt.show()
```
